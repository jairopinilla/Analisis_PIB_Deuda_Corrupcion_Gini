```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn import preprocessing 
from sklearn.metrics import pairwise_distances_argmin_min
import seaborn as sns
import warnings
from sklearn.ensemble import ExtraTreesClassifier


from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')

warnings.filterwarnings('ignore')
pd.set_option('display.max_rows', None)
```

```python
dataPIB_deuda_corrupcion = pd.read_excel('data_PIB_Corrupcion_Deuda.xlsx',index_col='PaisAgno',converters = {'Agno': str} )
dataPIB_deuda_corrupcion=dataPIB_deuda_corrupcion.dropna()
dataPIB_deuda_corrupcion.head()
```

```python
dataGini_Bruto = pd.read_excel('API_SI.POV.GINI_DS2_en_excel_v2_511456.xls')
dataGini_Bruto.head()
```

```python
def PivotYearBM(df, agnoinicio,agnofin,nombredato):
    lista=pd.DataFrame([['','','',0,0]], columns=['Country Name','Country Code','codeAgno','Agno',nombredato])
    for index, row in df.iterrows():
        for x in range(agnoinicio,agnofin+1):
            pais=row['Country Name']
            iso3=row['Country Code']
            agnostring=str(x)
            valor = row[agnostring]
            Agno=x
            codeAgno = iso3+'-'+str(x)
            listarow = pd.DataFrame([[pais,iso3,codeAgno,Agno,valor]],
                                    columns=['Country Name','Country Code','codeAgno','Agno',nombredato])
            lista=lista.append(listarow, ignore_index=True)
            #lista = numpy.append (lista, [pais, x, PaisAgno,valor])
    return lista
            
```

```python
dataGini=PivotYearBM(dataGini_Bruto,1995,2018,'GINI')
dataGini.head()
```

```python
dataGini=dataGini.dropna()
dataGini.head(20)
```

```python
#unificamos los dataset
dataset = pd.merge(dataGini, dataPIB_deuda_corrupcion, left_on='codeAgno',right_on='codeAgno',how='inner') 
dataset = dataset.dropna()
dataset = dataset.reset_index()
dataset.head()
```

```python
dataset.loc[dataset['nombre'] == 'Chile']
```

```python
#revisamos matriz de correlacion
corrmat = dataset[['DeudaPorcentajePIB','GINI','Indice_Corrupcion','PIBPerCapita']].corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, cbar=True, annot=True, fmt='.2f', annot_kws={'size': 10}, vmax=.5, square=True);
```

```python
#matriz
import matplotlib.pyplot as plt
sns.set()
cols = ['DeudaPorcentajePIB','GINI','Indice_Corrupcion','PIBPerCapita']
#sns.pairplot(dataset[cols], size = 15)
sns.pairplot(dataset[cols])
plt.show();
```

```python
#normalizar el los datos
dataPreAnalisis = dataset[['DeudaPorcentajePIB','GINI','Indice_Corrupcion','PIBPerCapita']]
#datapaisesPreAnalisis = datapaises[['corrupcionIndex','deuda','Economic freedon  score','DesempleoPorcentaje','PIBPerCapita']]
dataPreAnalisis.head()

min_max_scaler = preprocessing.MinMaxScaler() 
dataPreAnalisis_escalado = pd.DataFrame(min_max_scaler.fit_transform(dataPreAnalisis ),columns=dataPreAnalisis.columns, index=dataPreAnalisis.index) # Hay que convertir a DF el resultado.
dataPreAnalisis_escalado.head()
```

```python
x = dataPreAnalisis_escalado['PIBPerCapita'].values
y = dataPreAnalisis_escalado['GINI'].values
plt.xlabel('PIB per Capita')
plt.ylabel('GINI')
plt.title('PIB per Capita vs. GINI')
plt.plot(x,y,'o',markersize=5)
```

```python
x = dataPreAnalisis_escalado['PIBPerCapita'].values
y = dataPreAnalisis_escalado['Indice_Corrupcion'].values
plt.xlabel('PIB per Capita')
plt.ylabel('Indice_Corrupcion')
plt.title('PIB per Capita vs. Indice_Corrupcion')
plt.plot(x,y,'o',markersize=5)
```

```python
x = dataPreAnalisis_escalado['Indice_Corrupcion'].values
y = dataPreAnalisis_escalado['GINI'].values
plt.xlabel('Indice_Corrupcion')
plt.ylabel('GINI')
plt.title('Indice_Corrupcion vs. GINI')
plt.plot(x,y,'o',markersize=5)
```

```python
x = dataPreAnalisis_escalado['Indice_Corrupcion'].values
y = dataPreAnalisis_escalado['DeudaPorcentajePIB'].values
plt.xlabel('Indice_Corrupcion')
plt.ylabel('DeudaPorcentajePIB')
plt.title('Indice_Corrupcion vs. Deuda Porcentaje PIB')
plt.plot(x,y,'o',markersize=5)
```

```python
#varificamos cantidad de cluster
nc = range(1, 30)
kmeans = [KMeans(n_clusters=i) for i in nc]
score = [kmeans[i].fit(dataPreAnalisis_escalado).score(dataPreAnalisis_escalado) for i in range(len(kmeans))]
score
plt.xlabel('Cluster')
plt.ylabel('Errores Cuadraticos')
plt.plot(nc,score)
```

```python
#tomamos 5 cluster
kmeans = KMeans(n_clusters=5).fit(dataPreAnalisis_escalado)
labels = kmeans.predict(dataPreAnalisis_escalado)
dataset['Categoria'] = labels
dataset.head(20)
```

```python
dataset.loc[dataset['nombre'] == 'Chile']
```

```python
# cargar data
names =['DeudaPorcentajePIB','GINI','Indice_Corrupcion','PIBPerCapita']
array = dataPreAnalisis_escalado.values
X = array
Y = dataset[['Categoria']]

model = ExtraTreesClassifier(n_estimators=15)
model.fit(X, Y)
print(model.feature_importances_)

data=model.feature_importances_

pd.DataFrame(data, index=names)
```
